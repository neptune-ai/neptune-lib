# Overview

This library lets you download experiment data from neptune.ml.

Some useful features are:

**Get neptune.ml experiment view**

You can fetch neptune leaderboard to your local machine.

```python
leaderboard = project.get_leaderboard(state=['succeeded','aborted'])
leaderboard.head()
```

**Get numerical channel values**

You can fetch numerical channel values for any experiment.

```python
channel_df = experiment.get_numeric_channels_values('network_1 batch sum loss', 'network_5 batch sum loss')
channel_df.head()
```

**Get hardware utilization**

You can fetch hardware utilization like `mem/cpu/gpu` for any experiment.

```python
hardware_df = experiment.get_hardware_utilization()
hardware_df.head()
```
